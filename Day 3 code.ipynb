```
import pandas as pd
import statistics
from scipy.stats import skew

# Creating the dataset
data = {
    "Number": [1, 2, 3, 4, 5],
    "Age": [56, 46, 32, 60, 38],
    "Income": [61350.84215, 53777.18224, 39460.32263, 66672.12210, 44459.08553],
    "Spending_Score": [12372.86445, 11001.60423, 8007.385018, 13526.54837, 9059.304083],
    "Membership_Years": [15, 10, 19, 12, 9],
    "Purchase_Frequency": [77.685590, 51.858351, 98.166371, 62.530976, 46.470533],
    "Last_Purchase_Amount": [6232.12244, 5545.849698, 4054.645293, 6815.544393, 4617.833484],
    "Category": ["Gold", "Silver", "Gold", "Platinum", "Silver"],  # Nominal Data
    "Satisfaction_Level": [3, 2, 5, 4, 1],  # Ordinal Data
}

# Convert the data to a DataFrame
df = pd.DataFrame(data)

# Display the dataset
print("Dataset:")
print(df)

# Calculations
def calculate_statistics(column_name):
    col_data = df[column_name]
    mean_value = statistics.mean(col_data)
    median_value = statistics.median(col_data)
    try:
        mode_value = statistics.mode(col_data)
    except statistics.StatisticsError:
        mode_value = "No mode (all values unique)"
    range_value = max(col_data) - min(col_data)
    skewness_value = skew(col_data)
    variance_value = statistics.variance(col_data)
    std_deviation = statistics.stdev(col_data)

    # Detect outliers using 1.5 * IQR rule
    q1 = col_data.quantile(0.25) if hasattr(col_data, "quantile") else None
    q3 = col_data.quantile(0.75) if hasattr(col_data, "quantile") else None
    iqr = q3 - q1 if q1 is not None and q3 is not None else None
    outliers = col_data[(col_data < (q1 - 1.5 * iqr)) | (col_data > (q3 + 1.5 * iqr))] if iqr else []

    return {
        "Mean": mean_value,
        "Median": median_value,
        "Mode": mode_value,
        "Range": range_value,
        "Skewness": skewness_value,
        "Variance": variance_value,
        "Standard Deviation": std_deviation,
        "Outliers": list(outliers) if len(outliers) > 0 else "No outliers"
    }

# Display nominal and ordinal data
print("\nNominal Data (Categories):")
print(df["Category"].unique())

print("\nOrdinal Data (Satisfaction Levels):")
print(sorted(df["Satisfaction_Level"].unique()))

# Choose numerical columns for analysis
columns_to_analyze = ["Age", "Income", "Spending_Score", "Membership_Years", "Purchase_Frequency", "Last_Purchase_Amount"]

# Compute and display results
print("\nStatistics:")
for column in columns_to_analyze:
    stats = calculate_statistics(column)
    print(f"{column}: {stats}")
```
