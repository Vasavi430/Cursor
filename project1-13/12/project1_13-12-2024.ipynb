TASK 05 - MAKING FEATURE ENGINEERING. EXTRACTING NEW COLUMNS (DAY OF WEEK, DAY OF MONTH, HOUR, MONTH, SEASON, YEAR ETC.

TASK 09 - PLOTTING BIKE SHARES BY MONTHS & YEAR_OF_MONTH (USE LINEPLOT, POINTPOT, BARPLOT)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace with your actual data)
data = {'timestamp': ['2023-11-22 12:34:56', '2023-12-01 15:23:10', '2023-11-28 09:45:32'],
        'bike_shares': [100, 150, 80]}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert timestamp to datetime and set as index
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', inplace=True)

# Extract month and year-month
df['month'] = df.index.month
df['year_month'] = df.index.strftime('%Y-%m')

# Line plot by month
plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='bike_shares', data=df)
plt.title('Bike Shares by Month')
plt.xlabel('Month')
plt.ylabel('Bike Shares')
plt.show()

# Point plot by year-month
plt.figure(figsize=(10, 6))
sns.pointplot(x='year_month', y='bike_shares', data=df)
plt.title('Bike Shares by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.show()

# Bar plot by year-month
plt.figure(figsize=(10, 6))
sns.barplot(x='year_month', y='bike_shares', data=df)
plt.title('Bike Shares by Year-Month')
plt.xlabel('Year-Month')
plt.ylabel('Bike Shares')
plt.xticks(rotation=45)
plt.show()
```

TASK 13 - PLOTTING BIKE SHARES BY YEAR & PLOTTING BIKE SHARES ON HOLIDAYS BY SEASON

```python
# prompt: TASK 13 - PLOTTING BIKE SHARES BY YEAR & PLOTTING BIKE SHARES ON HOLIDAYS BY SEASON

# Sample data (replace with your actual data)
data = {'timestamp': pd.to_datetime(['2022-11-22 12:34:56', '2023-12-01 15:23:10', '2022-11-28 09:45:32', '2023-12-25 10:00:00', '2023-03-15 14:30:00']),
        'bike_shares': [100, 150, 80, 50, 120],
        'holiday': [0, 0, 0, 1, 0],  # 0 for non-holiday, 1 for holiday
        'season': [4, 1, 4, 1, 1]}

# Create a DataFrame
df = pd.DataFrame(data)
df = df.set_index('timestamp')

# Extract year
df['year'] = df.index.year

# Plot bike shares by year
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='bike_shares', data=df)
plt.title('Bike Shares by Year')
plt.xlabel('Year')
plt.ylabel('Bike Shares')
plt.show()

# Plot bike shares on holidays by season
holiday_data = df[df['holiday'] == 1]
plt.figure(figsize=(10,6))
sns.barplot(x='season', y='bike_shares', data = holiday_data)
plt.title('Bike Shares on Holidays by Season')
plt.xlabel('Season')
plt.ylabel('Bike Shares')
plt.show()
```

TASK 14 - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

```python
# prompt: TASK 14 - VISUALIZING THE DISTRIBUTION OF BIKE SHARES BY WEEKDAY-WEEKEND WITH USING PIECHART & BARPLOT

# Assuming 'df' is your DataFrame with 'timestamp' as the index and 'bike_shares' column

# Extract day of the week (0=Monday, 6=Sunday)
df['day_of_week'] = df.index.dayofweek

# Categorize as weekday (0-4) or weekend (5-6)
df['day_type'] = ['Weekend' if day in [5, 6] else 'Weekday' for day in df['day_of_week']]

# Calculate the total bike shares for each day type
bike_shares_by_day_type = df.groupby('day_type')['bike_shares'].sum()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(bike_shares_by_day_type, labels=bike_shares_by_day_type.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Bike Shares by Day Type (Pie Chart)')
plt.show()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=bike_shares_by_day_type.index, y=bike_shares_by_day_type.values)
plt.title('Distribution of Bike Shares by Day Type (Bar Plot)')
plt.xlabel('Day Type')
plt.ylabel('Total Bike Shares')
plt.show()
```
