TASK 14 :- Create categories for cars based on price (e.g., low, mid, high)

```python
# prompt: TASK 14 :- Create categories for cars based on price (e.g., low, mid, high)

# Define price thresholds for categories
price_thresholds = {
    'low': df['price'].quantile(0.33),  # 33rd percentile as the upper bound for 'low'
    'mid': df['price'].quantile(0.66)   # 66th percentile as the upper bound for 'mid'
}

# Function to categorize cars based on price
def categorize_price(price):
    if price <= price_thresholds['low']:
        return 'low'
    elif price <= price_thresholds['mid']:
        return 'mid'
    else:
        return 'high'

# Create the 'price_category' column using the categorize_price function
df['price_category'] = df['price'].apply(categorize_price)

# Display the DataFrame with the new 'price_category' column
df
```

TASK 19 :- Group the vehicles based on attributes like price, km, and age

```python
# prompt: TASK 19 :- Group the vehicles based on attributes like price, km, and age

import pandas as pd
# Group vehicles based on price, km, and age categories

# Define price, km, and age categories (adjust thresholds as needed)
price_bins = [0, 10000, 20000, 30000, float('inf')]
price_labels = ['Low', 'Medium', 'High', 'Very High']
km_bins = [0, 50000, 100000, 150000, float('inf')]
km_labels = ['Low', 'Medium', 'High', 'Very High']
age_bins = [0, 2, 5, 10, float('inf')]
age_labels = ['New', 'Young', 'Medium', 'Old']

# Create the categorical columns
df['price_group'] = pd.cut(df['price'], bins=price_bins, labels=price_labels, right=False)
df['km_group'] = pd.cut(df['km'], bins=km_bins, labels=km_labels, right=False)
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

# Group by the categorical columns and get the count of vehicles in each group
grouped_vehicles = df.groupby(['price_group', 'km_group', 'age_group']).size().reset_index(name='count')

# Display the grouped vehicles
print("\nGrouped Vehicles:")
print(grouped_vehicles)

# You can further analyze these groups. For instance, calculate the average price within each group
grouped_avg_price = df.groupby(['price_group', 'km_group', 'age_group'])['price'].mean().reset_index(name='average_price')
print("\nAverage Price within Groups:")
grouped_avg_price
```
