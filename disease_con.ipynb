Business problem:

The code assesses the patient's health status by analyzing model predictions, determining whether they were classified as "Healthy" or "Disease" based on the confusion matrix and evaluation metrics.


Summary

 1)Converts labels to numbers

2)Computes confusion matrix

3)Calculates accuracy & precision

4) Prints results

5)Visualizes confusion matrix

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score

# Load the dataset
file_path = '/content/confusion_matrix_dataset.csv' # Update this path if needed
df = pd.read_csv(file_path)

# Convert 'Disease' to 1 and 'Healthy' to 0
y_actual = np.where(df["Actual"] == "Disease", 1, 0)
y_pred = np.where(df["Predicted"] == "Disease", 1, 0)

# Compute confusion matrix
tn, fp, fn, tp = confusion_matrix(y_actual, y_pred).ravel()

# Calculate accuracy and precision
accuracy = accuracy_score(y_actual, y_pred)
precision = precision_score(y_actual, y_pred)

# Display results
print(f"True Positives (TP): {tp}")
print(f"True Negatives (TN): {tn}")
print(f"False Positives (FP): {fp}")
print(f"False Negatives (FN): {fn}")
print(f"Accuracy: {accuracy:.2%}")
print(f"Precision: {precision:.2%}")

# Create confusion matrix for visualization
conf_matrix = pd.DataFrame(
    [[tp, fn], [fp, tn]],
    index=["Actual Disease", "Actual Healthy"],
    columns=["Predicted Disease", "Predicted Healthy"]
)

# Plot confusion matrix
plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", linewidths=0.5)
plt.title("Confusion Matrix")
plt.ylabel("Actual Label")
plt.xlabel("Predicted Label")
plt.show()
```
